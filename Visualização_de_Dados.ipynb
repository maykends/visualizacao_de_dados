# Visualização de Dados

Aqui está um exemplo básico de um projeto de visualização de dados em Machine Learning usando Python. Vamos utilizar as bibliotecas **matplotlib** e **seaborn** para criar os gráficos.

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# Exemplo de dados

```python
idade = [25, 30, 35, 40, 45, 50, 55, 60, 65, 70]
salario = [5000, 6000, 7500, 8000, 9000, 9500, 10000, 11000, 12000, 13000]
```

# Gráfico de dispersão (scatter plot)

```python
plt.scatter(idade, salario)
plt.title("Relação entre Idade e Salário")
plt.xlabel("Idade")
plt.ylabel("Salário")
plt.show()
```

# Histograma

```python
plt.hist(salario, bins=5)
plt.title("Distribuição de Salários")
plt.xlabel("Salário")
plt.ylabel("Frequência")
plt.show()
```

# Gráfico de barras

```python
cargos = ['Engenheiro', 'Analista', 'Gerente', 'Diretor']
quantidade = [15, 10, 7, 3]
```

```python
plt.bar(cargos, quantidade)
plt.title("Quantidade de Funcionários por Cargo")
plt.xlabel("Cargo")
plt.ylabel("Quantidade")
plt.show()
```

# Gráfico de pizza

```python
vendas = [30, 20, 15, 10, 25]
produtos = ['Produto A', 'Produto B', 'Produto C', 'Produto D', 'Produto E']
```

```python
plt.pie(vendas, labels=produtos, autopct='%1.1f%%')
plt.title("Participação de Vendas por Produto")
plt.show()
```

# Gráfico de linhas

```python
anos = [2015, 2016, 2017, 2018, 2019, 2020]
receita = [10000, 15000, 12000, 18000, 20000, 25000]
```

```python
plt.plot(anos, receita)
plt.title("Receita ao longo dos anos")
plt.xlabel("Ano")
plt.ylabel("Receita")
plt.show()
```
